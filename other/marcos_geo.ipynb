```python
l = [31595,3285,2581,13155,2487,3843,2537,14912,10374,5482,6076,4489,1235,12240,10599,4426,691,5631,2173,1729,12134,3248,4940,4322,5000]
```

```python
l = target[['PROD']]

l
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = pd.read_csv(r'./preprocess.csv')
y = l
```

```python
SEED = 42

X_train, X_test, y_train, y_test= train_test_split(
    
    X,y,
    train_size   = 0.8,
    random_state = SEED,
)
```

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(n_estimators=3, max_depth=2, random_state=SEED)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(y_pred)
```

```python
import numpy as np
from sklearn import metrics

print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

## Define Grid 
grid = { 
    'n_estimators': [5,10,25],
    'max_features': ['sqrt','log2'],
    'max_depth' : [3,4,5,6,7],
    'random_state' : [SEED]
}## show start time

CV_rfr = GridSearchCV(estimator=RandomForestRegressor(), param_grid=grid, cv= 5)

CV_rfr.fit(X_train, y_train)## show end time

CV_rfr
```
